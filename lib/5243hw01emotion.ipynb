```python
sfs = {}
names = df.school.unique()
i = 0
ratios = []
for name in names:
    print(name)
    school_df = sen[sen['school'] == name]
    #all puntuations number
    all_punc_count = 0
    exclam_punc_count = 0
    
    for entry in school_df.iterrows():  # Iterate over all entries
        for char in entry[1].sentence_lowered:  # Iterate over all characters
            if char in ('!', "," ,"\'" ,";" ,"\"", ".", "-" ,"?"):  
                all_punc_count = all_punc_count + 1;
            if char in ('!', "?"):  
                exclam_punc_count = exclam_punc_count + 1;       
          
    print ("Total number of punctuation characters exists in ", name);  
    print (all_punc_count);  

    print ("Total number of exclamation punctuation characters exists in ", name);  
    print (exclam_punc_count);  
    
    print("exclamation ratio:")
    print(exclam_punc_count / all_punc_count)
    ratios.append(exclam_punc_count / all_punc_count)
```

```python

plt.rcParams["figure.figsize"] = (20,10)
plt.bar(names, ratios)
```
