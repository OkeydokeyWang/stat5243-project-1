## Part I：Harmony and Collisions between Philosophical Schools

Data Pre-processing
It would be a bit overwhelming to analyze the data directly through raw text so there were a couple steps we needed to take before getting into data analysis. First, we would need to tokenize all of the sentences to words, and summarize the general properties of the dataset. Then, we filter out the stop-words (words that doesn’t carry a specific meaning such as a, the etc.) and we will be only taking adjectives to show the general tone of all the work.
 
Data Analysis
Word cloud, word frequency

1. We can see that Plato's words are broad, abstract, without more specific division of objects within  humans. The words indeed connect with his philosophy type: abstract and utopian. Aristotle's words have more words such as “contrast”, “certain” and other words on the basis of being highly similar to Plato, giving us a more practical sense. His word cloud also includes words such as “animal” and “natural”, which begin to be in line with the later philosophical thoughts. Plato and Aristotle, as pioneers of philosophical thoughts, make us  more curious whether these roots correspond to rationalism or empiricism.

2. Rationalism is based on the claim that reason is the source of knowledge, while empiricism believes experience is far more crucial. From the word cloud figures, we can see that “evidence”, “certain”, “present” are shown in empiricism, which relate to experience. Rationalism has a larger content of words such as “necessary”, “true”, which embody the relationships between rationalism and truth. Besides, it is interesting that rationalism includes the word “god”. We understand that in this philosophical thought, "no evidence for any arbitrary supernatural authority e.g. God or Gods". The discussion of the existence of God may have become an integral part of rationalism.

3. The word cloud of the school called analytic philosophy shows words such as “true”, “certain”, “particular” and “logical”. They are consistent with the performance of this philosophical thought — to accurately evaluate every entity through natural science and logical thinking. 
The word cloud of continental philosophy includes “social”, “natural”, and “classical”, which further reflects the complex and abstract analysis of entities, especially the use of history.

4. In fact, both phenomenology and German idealism belong to continental philosophy. Phenomenology’s word cloud includes words such as “present”, “general”, “possible”, “object”, and “subject”.  They are related to the meaning of this philosophy itself — a deeper understanding of nature through analyzing daily human behavior. On the other hand, German idealism divides the world into internal and external parts. The view that the external can be analyzed but the internal cannot is also closely related to the words “general”, “possible”, “universal”, and “external” in its word cloud.

5. The emphasis on “labor”, “social” and “whole” in communism is very appropriate to its own meaning. In capitalism, “great”, “greater”, “different” and “whole” also belong to its doctrine . We could compare the difference between the two through these words. 
From the word frequency, the number of “labor” shown in communism is far ahead, which reflects the cornerstone of communism.

6. Stoicism is an original philosophy thought in ancient Greece and the Roman Empire. Its age supports why there are a lot of Early Modern English second person singular pronouns in its word cloud. In addition, we can also observe words such as “whole”, “natural”, “true”, and “universe”. They reflect the stoicism thought that "God, nature, and man are one, and one man is a universe".

7. Nietzsche's word cloud shows some other information besides his philosophical ideas. 
“German”: He is a German philosopher.
Christian: His later thoughts contained mainly fierce criticisms of Christianity.

8. The word cloud of feminism is very accurate: “young”, “black”, “sexual”, “female”, and “male”. These words all demonstrate againstment of gender stereotypes and the thought of equality with men. 
The frequency of “black” in Word frequency is relatively high, which makes me wonder whether the word represents a race or the emotion. After taking out the sentences containing “black”, we found that black not only appears with “woman”, but also with “Jewish”. Therefore, most of the time it represents the race and the disadvantaged groups at that time. In addition, it also appears as an adjective in the text, expressing the darkness of the oppression the women received.
